## **PREGUNTA 2**
¿Cómo co-varía cnt con temp, hum y windspeed? ¿La relación es lineal o solo monótona?. Ayuda: Calcular la correlación de Pearson y de Spearman.

```python
# Construimos nuevas columnas
columnas_relacion = ['cnt', 'temp', 'hum', 'windspeed']

# 2. Calcular correlaciones
columnas_relacion = ['cnt', 'temp', 'hum', 'windspeed']
pearson_corr = df[columnas_relacion].corr(method='pearson')
spearman_corr = df[columnas_relacion].corr(method='spearman')

print("Correlación de Pearson con cnt:")
print(pearson_corr)

print("\nCorrelación de Spearman con cnt:")
print(spearman_corr)
```

### **Interpretación pregunta 2**

Los resultados de las correlaciones de Pearson y Spearman nos ayudan a entender la relación entre la demanda diaria de bicicletas (cnt) y las variables climáticas **temp**, **hum** y **windspeed**.

**Correlación de Pearson:** Esta medida evalúa la relación lineal entre variables.

- **temp (temperatura):** La correlación de Pearson con **cnt** es positiva y moderada (aproximadamente 0.40). Esto sugiere una relación lineal positiva: a medida que aumenta la temperatura, la demanda de bicicletas tiende a aumentar.
- **hum (humedad):** La correlación de Pearson con **cnt** es negativa y moderada (aproximadamente -0.32). Esto indica una relación lineal negativa: a medida que aumenta la humedad, la demanda de bicicletas tiende a disminuir.
- **windspeed (velocidad del viento):** La correlación de Pearson con **cnt** es positiva y muy débil (aproximadamente 0.09). Esto sugiere una relación lineal positiva muy débil: a medida que aumenta la velocidad del viento, la demanda de bicicletas tiende a aumentar de forma muy leve.

**Correlación de Spearman:** Esta medida evalúa la relación monótona (si una variable tiende a aumentar o disminuir a medida que la otra lo hace, sin importar si la relación es lineal).

- **temp (temperatura):** La correlación de Spearman con **temp** es positiva y moderada (aproximadamente 0.42), similar a la de Pearson. Esto refuerza la idea de una relación monótona positiva: a mayor temperatura, mayor demanda.
- **hum (humedad):** La correlación de Spearman con **hum** es negativa y moderada (aproximadamente -0.36), similar a la de Pearson. Esto confirma una relación monótona negativa: a mayor humedad, menor demanda.
- **windspeed (velocidad del viento):** La correlación de Spearman con **windspeed** es positiva y débil (aproximadamente 0.13), slightly higher than Pearson's. This validates a weak positive monotonic relationship: a greater windspeed, a slightly higher demand.

**Conclusiones:**

- La temperatura tiene la relación positiva más fuerte con la demanda (cnt), y esta relación parece ser aproximadamente lineal y monótona.
- La humedad (hum) tiene una relación negativa moderada con la demanda, which also appears to be both linear and monotonic.
- La velocidad del viento (windspeed) has a very weak positive relationship with demand. Both Pearson and Spearman correlations are close to zero, indicating a negligible linear or monotonic relationship.

En resumen, el clima, especialmente la temperatura y la humedad, varía con la demanda de bicicletas. La relación con la temperatura es la más significativa y se puede considerar tanto lineal como monótona positiva. La humedad tiene una relación negativa moderada, también lineal y monótona. La velocidad del viento tiene un impacto mínimo y una relación muy débil con la demanda.

## **PREGUNTA 3**
¿Desde qué tamaño muestral el TLC justifica intervalos y pruebas basados en normalidad para la media de cnt?

```python
# Question 3: Central Limit Theorem and sample size

# Define a range of sample sizes to test
sample_sizes = [30, 50, 100, 200, 500]
n_simulations = 1000  # Number of sample means to generate for each sample size

plt.figure(figsize=(15, 10))

for i, sample_size in enumerate(sample_sizes):
    sample_means = []
    for _ in range(n_simulations):
        # Draw a random sample with replacement
        sample = df['cnt'].sample(sample_size, replace=True, random_state=rng)
        sample_means.append(sample.mean())

    plt.subplot(2, 3, i + 1)
    sns.histplot(sample_means, kde=True)
    plt.title(f'Sample Size: {sample_size}')
    plt.xlabel('Sample Mean of cnt')
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```


### **Interpretación pregunta 3**

Observando los histogramas generados para diferentes tamaños muestrales, podemos analizar cómo la distribución de la media muestral de cnt se aproxima a una distribución normal.

- **Tamaños muestrales pequeños:** La distribución de las medias muestrales aún puede mostrar cierta asimetría o desviación de la forma normal, reflejando en parte la distribución original de cnt.
- **Tamaños muestrales moderados a grandes:** A medida que el tamaño muestral aumenta, la distribución de las medias muestrales se vuelve progresivamente más simétrica y con forma de campana, asemejándose a una distribución normal.

Aunque no hay un tamaño muestral mágico universal, la regla general sugiere que para la mayoría de las distribuciones, un tamaño muestral de **n ≥ 30** es a menudo suficiente para que el TLC empiece a surtir efecto y la distribución de la media muestral sea aproximadamente normal.

En este caso particular, incluso con un tamaño muestral de 100, la distribución de las medias muestrales ya parece razonablemente normal. Para propósitos prácticos y para justificar el uso de intervalos de confianza y pruebas de hipótesis basados en la normalidad para la media de cnt, un tamaño muestral de **100 o más** parece adecuado según la visualización. Sin embargo, un tamaño muestral de **30** podría ser considerado el umbral mínimo donde el TLC comienza a ser relevante, aunque con una aproximación menos precisa que con tamaños mayores.

Por lo tanto, para la variable cnt, el TLC justifica el uso de métodos basados en normalidad para la media con tamaños muestrales de **alrededor de 100 o más**, aunque la aproximación es razonable a partir de **n=30**.

## **PREGUNTA 5**
¿Cuál es un IC bootstrap para la mediana de cnt y cómo cambia tu mensaje si la distribución es sesgada?

```python
# Número de remuestreos
n_iterations = 5000
sample_size = len(df['cnt'])
medians = []

for i in range(n_iterations):
    # Muestreo con reemplazo
    bootstrap_sample = df['cnt'].sample(sample_size, replace=True, random_state=rng)
    # Calcular la mediana de la muestra bootstrap
    median = bootstrap_sample.median()
    medians.append(median)

# Convert the list of medians to a pandas Series for plotting
medians_series = pd.Series(medians)

#calcular el intervalo del 95%
confidence_interval = np.percentile(medians_series, [2.5, 97.5])

print(f" bootstrap confidance Intervalo for the mediam of cnt: {confidence_interval}")

# calculate the mediam of the original date
original_median = df['cnt'].median()
print(f" original mediam of cnt: {original_median}")

plt.hist(medians_series, bins=50)
plt.title('Histogram of Bootstrap Medians')
plt.xlabel('Median of cnt')
plt.ylabel('Frequency')
plt.show()

sm.qqplot(medians_series, line='s')
plt.title('QQ Plot of Bootstrap Medians')
plt.show()
```

# Interpretación pregunta 5

Análisis y conclusión

- Un Intervalo de Confianza bootstrap para la mediana de cnt se construye mediante estos pasos:

Muestreo bootstrap: Tomar muchas muestras aleatorias con reemplazo del mismo tamaño que el conjunto original de cnt.

Cálculo de estadísticas: Calcular la mediana para cada muestra bootstrap.

es decir; se obtiene tomando percentiles (como 2.5 % y 97.5 %) de las medianas de múltiples muestras con reemplazo.

- Cuando la distribución de cnt es sesgada o (asimétrica): El intervalo de confianza bootstrap no será simétrico respecto a la mediana muestral. Esto no es un error, sino una característica útil: el IC bootstrap refleja la asimetría real de los datos.

- Con bootstrap:

El método de percentiles ya captura parte de la asimetría. Para mayor precisión, se puede usar el método BCa (bias-corrected and accelerated), que ajusta por sesgo y curvatura. Es decir; Si la distribución de cnt es sesgada, el intervalo también será asimétrico, y eso es correcto — refleja la realidad de los datos.
